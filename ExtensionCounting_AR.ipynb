```python
import pandas as pd
import os
```

```python
folder_path = '/home/brandonz/test/'
exts = ['filename', '--', 'NA', 'extA', 'extB', 'extC', 'extD', 'extE', 'extF', 'extG', 'extG', 'extH']
towrite = pd.DataFrame(columns = exts)

for filename in os.listdir(folder_path):
    filepath = os.path.join(folder_path, filename)
    
    if filename.endswith('.csv'):
        data = pd.read_csv(filepath, keep_default_na = False)
        counts = data['extension'].value_counts()
        newrow = {'filename': filename}
        newrow.update(counts.to_dict())
        towrite.loc[len(towrite)] = newrow

towrite = towrite.rename(columns = {'--': 'mutant', 'NA': 'unedited'})
towrite.fillna(0)
    
```
